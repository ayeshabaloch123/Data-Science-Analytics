```python
                      #Lab_7#


import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt


# Specify the path to your CSV file
csv_file_path = '/content/Diamonds Prices2022.csv'

# Use pandas to read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Now, you can work with the DataFrame 'df' as needed
# For example, you can display the first few rows of the DataFrame
print(df.head(20))

#The Unnamed: 0 column is of no use, so dropping that.

```

```python
#remove nan values
df.dropna(axis=0, inplace=True)
```

```python
df.head()
```

```python
# Bar Chart

sb.countplot(x="cut", data=df)
```

```python
# Pie Chart

df['color'].value_counts().plot.pie()
```

```python
# Histogram

plt.hist(data=df, x='price');
```

```python
# Scatter Plot

plt.scatter(x='carat', y='price', data=df)
plt.title("Scatter Plot: Carat vs Price")
plt.xlabel('Carat')
plt.ylabel('Price')
```

```python
# Heatmap

plt.hist2d(data=df, x='carat', y='price')
plt.colorbar()
plt.xlabel('Carat')
plt.ylabel('Price')
```

```python
# Boxplot

sb.boxplot(x='cut', y='price', data=df)
plt.xlabel('Cut')
plt.ylabel('Price')
```

```python
# Clustered Bar Chart (using Pandas)

cut_color_counts = df.groupby(['cut', 'color']).size().unstack()
cut_color_counts.plot(kind='bar', stacked=True)
plt.title("Clustered Bar Chart: Cut and Color Counts")
plt.xlabel("Cut")
plt.ylabel("Count")
```

```python
# Line Plot

sb.pointplot(data=df, x='cut', y='price')
plt.ylabel('Average Price')
plt.xlabel('Cut')
plt.xticks(rotation=15)
plt.title('Average Price by Cut')
```
